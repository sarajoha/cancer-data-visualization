```python
import pygal
```

```python
from IPython.display import display, SVG
```

```python
xvals = [0, 1, 3, 5, 6, 7, 9, 11, 12, 15]
```

```python
yvals = [4, 3, 1, 2, 2, 4, 5, 2, 1, 4]
```

```python
lineplot = pygal.Line(height=400)
```

```python
lineplot.title = 'My line plot'
```

```python
lineplot.x_labels = xvals
```

```python
lineplot.add('Data', yvals)
```

```python
display(SVG(lineplot.render(disable_xml_declaration=True)))
```
